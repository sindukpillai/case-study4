```python
%matplotlib inline
import pandas as pd
import statistics as st
import numpy as np
from matplotlib import pyplot as plt

data=pd.read_csv('C:\\Users\HP\Desktop\Sales_data_zone_wise.csv')
df=pd.DataFrame(data)

```

```python
#Average of Zone A
df1=df['Zone - A'].mean()
print(df1)

#Average of Zone B
df2=df['Zone - B'].mean()
print(df2)

#Average of Zone C
df3=df['Zone - C'].mean()
print(df3)

#Average of Zone D
df4=df['Zone - D'].mean()
print(df4)
```

```python
#Total sales generated for all the zones in each month.
df['Total Sales']=df['Zone - A'].sum()+df['Zone - B']+df['Zone - C']+df['Zone - D']
df
month_total=[df['Total Sales']]
plt.hist(month_total,30,color = "Green")
plt.show()
```

```python
#Total Sales for each zone
sales=[df['Zone - A'].sum(),df['Zone - B'].sum(),df['Zone - C'].sum(),df['Zone - D'].sum()]
sales

plt.show()
plt.hist(sales,10,color='skyblue')
```
